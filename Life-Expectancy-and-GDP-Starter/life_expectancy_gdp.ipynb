```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
all_data = pd.read_csv("all_data.csv")
data_germany = all_data[all_data.Country=="Germany"]
data_chile = all_data[all_data.Country=="Chile"]
data_china = all_data[all_data.Country=="China"]
data_mexico = all_data[all_data.Country=="Mexico"]
data_usa = all_data[all_data.Country=="United States of America"]
data_zimbabwe = all_data[all_data.Country=="Zimbabwe"]
```

```python
print(data_germany.head())
```

```python
#GDP of a country per year
ax1 = plt.subplot()
sns.lineplot(data=data_germany, x="Year", y="GDP")
ax1.set_xlabel("Year")
ax1.set_ylabel("GDP")
ax1.set_title("Germany")
plt.show()

ax2 = plt.subplot()
sns.lineplot(data=data_chile, x="Year", y="GDP")
ax2.set_xlabel("Year")
ax2.set_ylabel("GDP")
ax2.set_title("Chile")
plt.show()

ax3 = plt.subplot()
sns.lineplot(data=data_china, x="Year", y="GDP")
ax3.set_xlabel("Year")
ax3.set_ylabel("GDP")
ax3.set_title("China")
plt.show()

ax4 = plt.subplot()
sns.lineplot(data=data_mexico, x="Year", y="GDP")
ax4.set_xlabel("Year")
ax4.set_ylabel("GDP")
ax4.set_title("Mexico")
plt.show()

ax5 = plt.subplot()
sns.lineplot(data=data_usa, x="Year", y="GDP")
ax5.set_xlabel("Year")
ax5.set_ylabel("GDP")
ax5.set_title("USA")
plt.show()

ax6 = plt.subplot()
sns.lineplot(data=data_zimbabwe, x="Year", y="GDP")
ax6.set_xlabel("Year")
ax6.set_ylabel("GDP")
ax6.set_title("Zimbabwe")
plt.show()
```

```python
#Comparing life expectancies of various countries from 2000-2015
ax = plt.subplot()
sns.lineplot(data=data_germany, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("Germany")

ax = plt.subplot()
sns.lineplot(data=data_chile, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("Chile")

ax = plt.subplot()
sns.lineplot(data=data_china, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("China")

ax = plt.subplot()
sns.lineplot(data=data_mexico, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("Mexico")

ax = plt.subplot()
sns.lineplot(data=data_usa, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("USA")

ax = plt.subplot()
sns.lineplot(data=data_zimbabwe, x="Year", y="Life expectancy at birth (years)")
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy")
ax.set_title("Zimbabwe")

plt.legend(["Germany", "Chile", "China", "Mexico", "USA", "Zimbabwe"])
plt.show()
```

```python
def func(country):
    if country == 'United States of America':
        return 'USA'
    return country
all_data['Country'] = all_data.apply(lambda x: func(x['Country']), axis=1)
ax = plt.subplot()
sns.barplot(data=all_data, y="GDP", x="Country")
ax.set_ylabel("GDP")
ax.set_xlabel("Country")
ax.set_title("")
plt.show()
```
